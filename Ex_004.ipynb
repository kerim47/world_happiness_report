```python
import pandas as pd
import numpy as np

# Rastgele deerleri reten fonksiyonu tanmla
def random_arr():
    return [np.random.randint(1, 34) for _ in range(10)]

# Veri setini olutur
sns_data_1 = random_arr()
sns_data_2 = random_arr()

df = pd.DataFrame(
    {
        'column_1': sns_data_1,
        'column_2': sns_data_2
    },
    index=random_arr()
)

print(df)

# Dalm grafii oluturma
sns.jointplot(x='column_1', y='column_2', data=df)
# sns.lmplot(x='column_1', y='column_2', data=df)
# Dalm grafii oluturma
sns.lmplot(x='column_1', y='column_2', data=df)
```
